```python
import os
import requests

# Ваш API-ключ Jamendo
JAMENDO_CLIENT_ID = 'fe99e126'

def search_tracks(query):
    url = 'https://api.jamendo.com/v3.0/tracks/'
    params = {
        'client_id': JAMENDO_CLIENT_ID,
        'format': 'json',
        'namesearch': query,
        'limit': 10
    }

    response = requests.get(url, params=params)
    tracks = response.json().get('results', [])

    print("Найденные треки:")
    for i, track in enumerate(tracks, 1):
        print(f"{i}. {track['name']} - {track['artist_name']}")

    return tracks

def download_track(track, save_folder="downloads"):
    os.makedirs(save_folder, exist_ok=True)

    track_name = track['name']
    artist_name = track['artist_name']
    audio_url = track['audio']
    filename = f"{save_folder}/{track_name} - {artist_name}.mp3"

    # Скачивание трека
    response = requests.get(audio_url)
    with open(filename, 'wb') as f:
        f.write(response.content)

    print(f"Трек скачан: {filename}")

def search_albums(query):
    url = 'https://api.jamendo.com/v3.0/albums/'
    params = {
        'client_id': JAMENDO_CLIENT_ID,
        'format': 'json',
        'namesearch': query,
        'limit': 10
    }

    response = requests.get(url, params=params)
    albums = response.json().get('results', [])

    print("Найденные альбомы:")
    for i, album in enumerate(albums, 1):
        print(f"{i}. {album['name']} - {album['artist_name']}")

    return albums

def download_album(album_id, save_folder="downloads"):
    url = f'https://api.jamendo.com/v3.0/albums/tracks/?client_id={JAMENDO_CLIENT_ID}&format=json&id={album_id}'
    response = requests.get(url)
    tracks = response.json().get('results', [])

    album_folder = f"{save_folder}/{tracks[0]['album_name']}"
    os.makedirs(album_folder, exist_ok=True)

    for track in tracks:
        track_name = track['name']
        artist_name = track['artist_name']
        audio_url = track['audio']
        filename = f"{album_folder}/{track_name} - {artist_name}.mp3"

        # Скачивание трека
        response = requests.get(audio_url)
        with open(filename, 'wb') as f:
            f.write(response.content)

        print(f"Трек скачан: {filename}")

def search_artist(query):
    url = 'https://api.jamendo.com/v3.0/artists/'
    params = {
        'client_id': JAMENDO_CLIENT_ID,
        'format': 'json',
        'namesearch': query,
        'limit': 10
    }

    response = requests.get(url, params=params)
    artists = response.json().get('results', [])

    print("Найденные артисты:")
    for i, artist in enumerate(artists, 1):
        print(f"{i}. {artist['name']}")

    return artists

# Пример использования
while True:
    query = input("Введите название трека, альбома, артиста или 'артист - песня': ")

    # Попробуем найти треки
    tracks = search_tracks(query)
    choice = int(input("Введите номер трека для скачивания, 0 для поиска альбомов, -1 для поиска артистов или -2 для повторного поиска: "))

    if choice == -2:
        continue
    elif choice == -1:
        artists = search_artist(query)
        artist_choice = int(input("Введите номер артиста для поиска треков или альбомов, или 0 для повторного поиска: "))
        if artist_choice != 0:
            selected_artist = artists[artist_choice - 1]
            query = selected_artist['name']
            tracks = search_tracks(query)
            choice = int(input("Введите номер трека для скачивания или 0 для поиска альбомов: "))
            if choice != 0:
                selected_track = tracks[choice - 1]
                download_track(selected_track)
                break
            else:
                albums = search_albums(query)
                album_choice = int(input("Введите номер альбома для скачивания: "))
                selected_album = albums[album_choice - 1]
                download_album(selected_album['id'])
                break
    elif choice == 0:
        albums = search_albums(query)
        album_choice = int(input("Введите номер альбома для скачивания или 0 для повторного поиска: "))
        if album_choice != 0:
            selected_album = albums[album_choice - 1]
            download_album(selected_album['id'])
            break
    elif 1 <= choice <= len(tracks):
        selected_track = tracks[choice - 1]
        download_track(selected_track)
        break
    else:
        print("Неверный выбор. Пожалуйста, попробуйте снова.")
```
